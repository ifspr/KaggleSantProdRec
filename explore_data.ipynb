### Read in the training set

```python
df.ind_nomina_ult1.fillna(0.0,inplace=True) 
df.ind_nom_pens_ult1.fillna(0.0,inplace=True)
df.ind_nomina_ult1=df.ind_nomina_ult1.astype(int)
df.ind_nom_pens_ult1=df.ind_nom_pens_ult1.astype(int)
```

### Plot the number of customers owning a certain product as a function of time

```python
products = df.iloc[:1,].filter(regex="ind_+.*ult.*").columns.values
```

```python
df["fecha_dato"] = pd.to_datetime(df["fecha_dato"],format="%Y-%m-%d")
df2=df[products].groupby(df.fecha_dato).sum()
df2['yyyymm_dato']=pd.DatetimeIndex(df['fecha_dato'].unique()).year*100+\
                   pd.DatetimeIndex(df['fecha_dato'].unique()).month
ndates=len(df.fecha_dato.unique())

#plot the numbers of customers owning a certain product as a function of time
fig = plt.figure(figsize=(13,30))
plt.subplots_adjust(wspace=0.3, hspace=1.6)
for col_id in range(len(products)):
    ax = fig.add_subplot(12,2,col_id+1)
    ax.plot(range(ndates),df2[products[col_id]])
    ax.set_xticks(range(ndates)) 
    ax.set_xticklabels(df2['yyyymm_dato'])
    ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
    ax.set_title(products[col_id])
    ax.set_ylabel('count')
    if (col_id==22 or col_id==23): ax.set_xlabel('yyyymm')
```

### Plot the age distribution and how mean 'renta' varies with age

```python
renta_age = df.renta.groupby(pd.cut(df['age'],np.linspace(20,80,13))).mean()

fig, subaxes = plt.subplots(1, 2, figsize=(13,5))
plt.subplots_adjust(wspace=0.4)
subaxes[0].plot(np.linspace(22.5,77.5,12),renta_age)
subaxes[0].set_xlabel('Age')
subaxes[0].set_ylabel('Renta')
sns.distplot(df.age[(df.age>=5) & (df.age<=120)], ax=subaxes[1], axlabel='Age', kde=False, bins=30)
subaxes[1].set_ylabel('Frequency')
```

### Plot the income distribution in each 'prov' and for each gender group

```python
df.cod_prov.fillna(0.0,inplace=True)
df.cod_prov=df.cod_prov.astype(int)

fig, subaxes = plt.subplots(4, 1, figsize=(13,20))
plt.subplots_adjust(hspace=1.0)
for ax, k, l in zip(subaxes, [1,14,27,40], [14,27,40,54]):
  dset = df.loc[(df['renta']<350000) & (df.cod_prov >= k) & (df.cod_prov < l),\
                ['renta','nomprov','sexo']]
  g=sns.violinplot(x="nomprov",y="renta",hue='sexo',data=dset,ax=ax,split=True)
  g.set_xticklabels(g.get_xticklabels(),rotation=30)
```
